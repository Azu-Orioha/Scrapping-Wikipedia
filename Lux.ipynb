```python
from bs4 import BeautifulSoup

def find_toc(soup):
    # Try multiple search methods
    selectors = [
        {'name': 'div', 'attrs': {'class': 'toc'}},
        {'name': 'div', 'attrs': {'class': 'table-of-contents'}},
        {'name': 'nav', 'attrs': {'aria-label': 'Table of Contents'}},
        {'css': 'div.toc'},
        {'css': '[class*="toc"]'},
    ]
    
    for selector in selectors:
        if 'css' in selector:
            toc = soup.select_one(selector['css'])
        else:
            toc = soup.find(**selector)
        
        if toc:
            return toc
    
    return None

toc_div = find_toc(soup)

if toc_div:
    print("Found TOC container!")
    links = toc_div.find_all('a')
    for link in links:
        print(link.text)
else:
    print("Could not find table of contents in the HTML")
    print("Available classes found:", {tag.get('class') for tag in soup.find_all(class_=True)})
```

```python
import requests
from bs4 import BeautifulSoup
import re

url = "https://en.wikipedia.org/wiki/Luxembourg"
headers = {'User-Agent': 'Mozilla/5.0'}

try:
    # Fetch the page with error handling
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    found_language = False
    

    # METHOD 2: Fallback - Search article paragraphs
    if not found_language:
        print("=== Method 2: Searching Article Text ===")
        paragraphs = soup.select('div.mw-parser-output > p')
        for para in paragraphs:
            text = para.get_text()
            if re.search(r'(official|national)\s+language', text, re.IGNORECASE):
                # Capture the sentence where languages are mentioned
                sentences = re.split(r'(?<=[.!?])\s+', text)
                for sentence in sentences:
                    if re.search(r'(official|national)\s+language', sentence, re.IGNORECASE):
                        print("Found in text:", sentence.strip())
                        found_language = True
                        break
            if found_language:
                break

    if not found_language:
        print("No language information found.")

except requests.RequestException as e:
    print(f"Request failed: {e}")

    
    
```
